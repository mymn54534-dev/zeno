### كود حساب درجات الطلاب

يأخذ هذا الكود درجات طالب في مواد مختلفة، يحسب متوسط الدرجات، ثم يحدد التقدير الحرفي بناءً على هذا المتوسط.

```python
# دالة لحساب متوسط الدرجات وتحديد التقدير
def calculate_student_grade(scores):
    if not scores:
        return "لا توجد درجات لإدخالها", "F"

    # حساب مجموع الدرجات
    total_score = sum(scores)
    # حساب متوسط الدرجات
    average_score = total_score / len(scores)

    # تحديد التقدير الحرفي بناءً على المتوسط
    if average_score >= 90:
        grade = "A"
    elif average_score >= 80:
        grade = "B"
    elif average_score >= 70:
        grade = "C"
    elif average_score >= 60:
        grade = "D"
    else:
        grade = "F"

    return average_score, grade

# مثال على استخدام الدالة:
# درجات الطالب (يمكنك تغيير هذه الدرجات)
student_scores = [85, 92, 78, 65, 90]

# حساب الدرجة والتقدير
average, grade = calculate_student_grade(student_scores)

# طباعة النتائج
print(f"درجات الطالب: {student_scores}")
print(f"متوسط الدرجات: {average:.2f}")
print(f"التقدير الحرفي: {grade}")

# مثال آخر
student_scores_2 = [55, 60, 45]
average_2, grade_2 = calculate_student_grade(student_scores_2)
print(f"\nدرجات الطالب 2: {student_scores_2}")
print(f"متوسط الدرجات 2: {average_2:.2f}")
print(f"التقدير الحرفي 2: {grade_2}")
```
